# Data Visualisation

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset("tips")
type(tips)
```

```python
tips.smoker.value_counts().plot(kind='bar', rot=0)
plt.title("Bar Plot")
plt.xlabel("Smokers")
plt.ylabel("Count")
# plt.show()
```
